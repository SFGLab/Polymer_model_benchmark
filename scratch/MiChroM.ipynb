## Data preparation

```python
import pandas as pd

def loops_to_txt(bedpe_file,N_beads,region,chrom,normalization=False,viz=False):
    # Read file and select the region of interest
    df = pd.read_csv(bedpe_file,sep='\t',header=None)
    df = df[(df[1]>=region[0])&(df[2]>=region[0])&(df[4]>=region[0])&(df[5]>=region[0])&(df[5]<region[1])&(df[4]<region[1])&(df[1]<region[1])&(df[2]<region[1])&(df[0]==chrom)].reset_index(drop=True)

    # Convert hic coords into simulation beads
    resolution = (region[1]-region[0])//N_beads
    df[1], df[2], df[4], df[5] = (df[1]-region[0])//resolution, (df[2]-region[0])//resolution, (df[4]-region[0])//resolution, (df[5]-region[0])//resolution
    
    content = ''
        
    for i in range(len(df)):
        content+=f'{int((df[1][i]+df[2][i])//2)+1} {int((df[4][i]+df[5][i])//2)+1}\n'

    f = open("loops.txt", "w")
    f.write(content)
    f.close()

chrom = 'chr1'
region = [178390648,179460328]
N_beads = 1000

bedpe_file = '/home/skorsak/Documents/data/Rao/GSE63525_GM12878_primary+replicate_HiCCUPS_looplist_hg19_2.bedpe'
loops_to_txt(bedpe_file,N_beads,region,chrom,normalization=False,viz=False)
```

```python
def compartments_to_txt(comp_file,N_beads,region,chrom,normalization=False,viz=False):
    # Read file and select the region of interest
    df = pd.read_csv(bedpe_file,sep='\t',header=None)
    df = df[(df[1]>=region[0])&(df[2]>=region[0])&(df[4]>=region[0])&(df[5]>=region[0])&(df[5]<region[1])&(df[4]<region[1])&(df[1]<region[1])&(df[2]<region[1])&(df[0]==chrom)].reset_index(drop=True)

    # Convert hic coords into simulation beads
    resolution = (region[1]-region[0])//N_beads
    df[1], df[2], df[4], df[5] = (df[1]-region[0])//resolution, (df[2]-region[0])//resolution, (df[4]-region[0])//resolution, (df[5]-region[0])//resolution
    
    content = ''
    
    for i in range(len(df)):
        content+=f'{int((df[1][i]+df[2][i])//2)} {int((df[4][i]+df[5][i])//2)}\n'

    f = open("loops.txt", "w")
    f.write(content)
    f.close()

comp_file = '/home/skorsak/Documents/data/Trios/calder_HiChIP_subcomp/CHS_d.bed'
df = pd.read_csv(comp_file,sep='\t',header=None)
df = df[((df[2]>=region[0])&(df[1]<region[1])&(df[0]==chrom))].reset_index(drop=True)

# Convert hic coords into simulation beads
resolution = (region[1]-region[0])//N_beads
df[1], df[2] = (df[1]-region[0])//resolution, (df[2]-region[0])//resolution

df[1][0] = 0
df[2][len(df)-1] = N_beads

content = ''
for i in range(len(df)):
    start = df[1][i]
    end = df[2][i]
    type = df[3][i]
    if type.startswith('A.1'):
        type = 'A1'
    elif type.startswith('A.2'):
        type = 'A2'
    elif type.startswith('A.3'):
        type = 'A3'
    elif type.startswith('B.1'):
        type = 'B1'
    elif type.startswith('B.2'):
        type = 'B2'
    elif type.startswith('B.3'):
        type = 'B3'

    for i in range(start,end):
        content+=f'{i+1} {type}\n'

f = open("compartments.txt", "w")
f.write(content)
f.close()
```
