```python
import cooler
import numpy as np
import matplotlib.pyplot as plt

# Load specific resolution (adjust if needed, e.g., 10kb bins)
cool_path = "/home/skorsak/Data/4DNucleome/SPRITE_4DNESI1U7ZW9/4DNFIUOOYQC3.mcool::/resolutions/5000"
clr = cooler.Cooler(cool_path)

# Define region (remove dots)
region = "chr1:178421513-179491193"

# Fetch the matrix (non-balanced)
mat = clr.matrix(balance=False).fetch(region)

# Convert to NumPy array
array = np.array(mat)

# Save as .npy file
np.save("sprite_chr1_178421513_179491193.npy", array)

# Plot for visualization
plt.imshow(array, cmap='Reds', interpolation='none',vmax=np.mean(array)+np.std(array))
plt.title("chr1:178421513179491193")
plt.colorbar(label='Contact count')
plt.tight_layout()
plt.savefig("sprite_chr1_178421513_179491193.png", dpi=300)
plt.show()
```
